```python
# result was generated using C:\Users\tlinden\Documents\Git\leoss\src\models\evaluate_models.py
results_ncv = pd.read_csv(r"C:\Users\tlinden\Documents\Git\leoss\results\nCV\metrics_nCV.csv", sep = "\t", index_col=0)
results_ncv
```

```python
d1 = cuno.reset_index().set_index(["index","model"])
d2 = results_ncv.reset_index().set_index(["index","model"])
d = d1.join(d2)
d
```

```python
d.groupby("model").agg(["mean","std"])
```

```python
# https://loft-br.github.io/xgboost-survival-embeddings/benchmarks/benchmarks.html

f = plt.figure()
f.set_figheight(8)
f.set_figwidth(8)

# Iterate through continents
cols = ["charrell","ctd","dcal_max","ibs","cuno"]
titles = ["Harrell's C","time-dependent concordance index","maximum decile deviation","IBS","Uno's C"]

for i in range(len(cols)):
    col = cols[i]
    coltitle = titles[i]
    # create subplot axes in a 3x3 grid
    ax = plt.subplot(2, 3, i + 1) # nrows, ncols, axes position
    # plot on axis
    d.boxplot(
        by="model", 
        column= col, 
        grid = True,
        figsize=(4,4), 
        ax = ax,
        )
    # set the title
    ax.set_title(coltitle) 

plt.tight_layout()
plt.suptitle("model comparison: nested cross-validation test-set performance")
# save
# plt.savefig("nCV_boxplot_modelcomparison.pdf", bbox_inches="tight")
plt.show()
```

# beautiful boxplot with seaborn

```python
d_long = pd.melt(d.reset_index().drop("index",axis=1), id_vars="model")
d_long
```

```python
# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 32}

# matplotlib.rc('font', **font)
```

```python
import seaborn as sns
# sns.set(font_scale=5)
cols = ["cuno","ibs"]
titles = ["Uno's C-Index","IBS"]
for i in range(len(cols)):
    data_ = d_long[d_long.variable==cols[i]]
    sns.set_theme(style="whitegrid")
    ax = sns.boxplot(
        x="model",
        y="value",    
        data=data_,
        labels=[""],
        palette="colorblind",
        fliersize=7.5,
        whis=4,
        linewidth=1,
        saturation=0.75,
        medianprops=dict(color="black", alpha=10)
        # hue="model",
        )    
    ax.set_yticklabels(np.round(ax.get_yticks(),2), size = 14)
    ax.set_xticklabels(ax.get_xmajorticklabels(), size = 14)
    if cols[i] == "cuno":
        ax.set_ylim(0.5,1)
    ax.set_ylabel("")
    ax.set_xlabel("\nmodel", fontsize=16)
    ax.set_title(titles[i], fontsize=20)
    plt.savefig(f"results_nCV_boxplot_{cols[i]}.pdf", bbox_inches="tight")
    plt.show()
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
